```python
def get_user_info():
    name = input("Enter your name: ")
    email = input("Enter your email: ")
    return {"name": name, "email": email}
def timing():
    a = int(input("choose your time: "))
    time1 = {
        "1": "10.00-1.00",
        "2": "1.10-4.10",
        "3": "4.20-7.20",
        "4": "7.30-10.30"
    }
    time2 = {
        "1": "10.15-1.15",
        "2": "1.25-4.25",
        "3": "4.35-7.35",
        "4": "7.45-10.45"
    }
    time3 = {
        "1": "10.30-1.30",
        "2": "1.40-4.40",
        "3": "4.50-7.50",
        "4": "8.00-10.45"
    }
    if a == 1:
        print("choose your time:")
        print(time1)
        t = input("select your time:")
        x = time1[t]
        print("successful!, enjoy movie at "+x)
    elif a == 2:
        print("choose your time:")
        print(time2)
        t = input("select your time:")
        x = time2[t]
        print("successful!, enjoy movie at "+x)
    return
def create_seating(rows, cols):
    seating = [['O' for _ in range(cols)] for _ in range(rows)]
    return seating

def display_seating(seating):
    for row in seating:
        print(" ".join(row))
    print()

def book_seat(seating, row, col):
    if seating[row][col] == 'O':
        seating[row][col] = 'X'
        print("Seat booked successfully!")
    else:
        print("Seat is already booked.")
def main():
    rows = int(input("Enter the number of rows: "))
    cols = int(input("Enter the number of columns: "))

    seating = create_seating(rows, cols)

    while True:
        print("\nCurrent Seating:")
        display_seating(seating)

        row = int(input("Enter the row number (starting from 0): "))
        col = int(input("Enter the column number (starting from 0): "))

        if 0 <= row < rows and 0 <= col < cols:
            book_seat(seating, row, col)
        else:
            print("Invalid row or column number. Please try again.")

        another_booking = input("Do you want to make another booking? (yes/no): ")
        if another_booking.lower() != "yes":
            break
def book_movieticketsscreening():
    print("\nWELCOME TO THE MOVIE TICKET BOOKING SYSTEM \n")
    
    movie={
        1:"DASARA",
        2:"SPY",
        3:"SEETHARAMAN"
    }
    print("Available movie:")
    for key,value in movie.items():
        print("{a},{b}".format(a=key,b=value))
    movie_choice=int(input("\nenter the movie number for booking:"))
    if movie_choice not in movie:
        print("invalid movie selection")
        return
    selected_movie=movie[movie_choice]
    num_tickets=int(input("enter the number ticket"))
    tickets_price=200
    totalprice=tickets_price*num_tickets
    user_info = get_user_info()
    t=timing()
    main()
    print("booking details")
    print(f"movies name : {selected_movie}\nnumber of tickets:{num_tickets}")
    print(f"totalprice :{totalprice}/-")
    print(f"User: {user_info['name']}")
    print(f"Email: {user_info['email']}")
    print(f"time:{t}")
    confirm=input("confirm booking?(yes/no):")
    if confirm.lower()=="yes":
        print("Tickets booked successfully!")
    else:
        print("booking canceled booking again,sorry")
    
book_movieticketsscreening()





        
```
